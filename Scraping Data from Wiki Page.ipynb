```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_total_number_of_speakers'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find('table', class_ = 'wikitable')
```

```python
table_columns = soup.find('table').find_all('th')
table_columns
```

```python
my_columns = [title.text.strip() for title in table_columns] 
my_columns
```

```python
df = pd.DataFrame(columns = my_columns)
df
```

```python
column_data = soup.find('table').find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    ind_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = ind_row_data
```
